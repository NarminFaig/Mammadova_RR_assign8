```python
import matplotlib.pyplot as plt
import pandas as pd

# Load the dataset
df = pd.read_csv("/Users/narminmammadova/Desktop/Game_of_Throness.csv")

# Calculate average viewership
average_viewership = df['US_Viewers_in_Millions'].mean()

# Find highest viewed episode
highest_viewed = df.loc[df['US_Viewers_in_Millions'].idxmax()]

# Find lowest viewed episode
lowest_viewed = df.loc[df['US_Viewers_in_Millions'].idxmin()]

# Display results
print("Average Viewership: {:.2f} million".format(average_viewership))
print("Highest Viewed Episode:")
print(highest_viewed)
print("Lowest Viewed Episode:")
print(lowest_viewed)

# Plotting
plt.figure(figsize=(10, 5))

# Plotting the viewership data
plt.subplot(1, 2, 1)
plt.bar(df['No_of_Episode_in_Overall'], df['US_Viewers_in_Millions'], color='darkblue')
plt.xlabel('No_of_Episode_in_Season')
plt.ylabel('Viewership (millions)')
plt.title('Viewership per Episode')

# Plotting average viewership
plt.subplot(1, 2, 2)
plt.bar(['Average'], [average_viewership], color='pink')
plt.xlabel('Viewership')
plt.ylabel('Average Viewership (millions)')
plt.title('Average Viewership: {:.2f} million'.format(average_viewership))

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Convert air date to datetime
df['Air_Date'] = pd.to_datetime(df['Air_Date'])

# Plot viewership over time
plt.figure(figsize=(10, 6))
plt.plot(df['Air_Date'], df['US_Viewers_in_Millions'], color='red')
plt.title('Viewership of Game_of_Thrones')
plt.xlabel('Year')
plt.ylabel('Viewers (millions)')
plt.grid(True)
plt.show()
```

```python
# Calculate episode-to-episode changes in viewership
df['viewership_change'] = df['US_Viewers_in_Millions'].diff()

# Plot episode-to-episode changes in viewership
plt.figure(figsize=(10, 6))
plt.plot(df['Air_Date'], df['viewership_change'], color='yellow')
plt.title('Episode-to-Episode Changes in Viewership of Game_of_Thrones')
plt.xlabel('Year')
plt.ylabel('Change in Viewership (millions)')
plt.grid(True)
plt.show()
```
